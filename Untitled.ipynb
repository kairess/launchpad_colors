```python
import matplotlib.pyplot as plt
import numpy as np
import cv2, colorsys
from sklearn.metrics import pairwise_distances_argmin
```

```python
colors_hex = [
    # 0..63
    "#000000", "#505050", "#a0a0a0", "#f0f0f0", "#ff9185", "#ff230a", "#e61f09", "#cc1c08",
    "#ffc285", "#ff850a", "#e67709", "#cc6a08", "#fff385", "#ffe70a", "#e6cf09", "#ccb808",
    "#daff85", "#b6ff0a", "#a3e609", "#91cc08", "#a9ff85", "#54ff0a", "#4be609", "#43cc08",
    "#85ff91", "#0aff23", "#09e61f", "#08cc1c", "#85ffc2", "#0aff85", "#09e677", "#08cc6a",
    "#85fff3", "#0affe7", "#09e6cf", "#08ccb8", "#85daff", "#0ab6ff", "#09a3e6", "#0891cc",
    "#85a9ff", "#0a54ff", "#094be6", "#0843cc", "#9185ff", "#230aff", "#1f09e6", "#1c08cc",
    "#c285ff", "#850aff", "#7709e6", "#6a08cc", "#f385ff", "#e70aff", "#cf09e6", "#b808cc",
    "#ff85da", "#ff0ab6", "#e609a3", "#cc0891", "#f04115", "#bf6100", "#b18c00", "#859708",
    # 64..127
    "#50a027", "#009d8e", "#0079c0", "#0000ff", "#2d50a4", "#6247b0", "#7b7b7b", "#3c3c3c",
    "#ff0000", "#bfbb64", "#a6c000", "#78c823", "#34c500", "#00c0af", "#00a2f1", "#527de7",
    "#8868e7", "#a447af", "#b93b69", "#975731", "#f86c00", "#befd00", "#82ff5d", "#00ff00",
    "#00ffa5", "#52ffe8", "#00e9ff", "#89c4ff", "#91a5ff", "#b989ff", "#da67e7", "#ff2cd6",
    "#ffa601", "#fff200", "#e3f600", "#dcc500", "#bf9e5f", "#88b57b", "#86c2ba", "#9ab3c5",
    "#84a5c3", "#c78b7a", "#f43c7f", "#ff93a5", "#ffa36f", "#ffef9a", "#d2e594", "#bad16f",
    "#a9a9a9", "#d3fee0", "#ccf1f9", "#b9c0e4", "#cdbae5", "#d0d0d0", "#dfe6e5", "#ffffff",
    "#f2210a", "#bf1a08", "#acf20a", "#88bf08", "#f2db0a", "#bfad08", "#f7a409", "#e06b41",
]
```

```python
colors_rgb = []

for i, h in enumerate(colors_hex):
    h = h.lstrip('#')
    rgb = tuple(int(h[i:i+2], 16) for i in (0, 2 ,4))
    
    colors_rgb.append(rgb)
    
    pixel = np.repeat(np.reshape(rgb, (1, 1, 3)), 10, axis=0)
    pixel = np.repeat(pixel, 10, axis=1)

    plt.subplot(8, 16, i+1)
    plt.title(rgb)
    plt.imshow(pixel)
```

```python
img = cv2.imread('00.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, dsize=None, fx=0.1, fy=0.1, interpolation=cv2.INTER_NEAREST)

print(img.shape)

plt.imshow(img)
```

```python
new_img_color_idx = pairwise_distances_argmin(img.reshape((-1, 3)), colors_rgb).reshape(img.shape[:2])

new_img = np.zeros_like(img, dtype=np.uint8)

for y in range(img.shape[0]):
    for x in range(img.shape[1]):
        color = colors_rgb[new_img_color_idx[y, x]]
        
        new_img[y, x] = color
        
plt.imshow(new_img)
```

```python
colors_hsv = []

for i, (r, g, b) in enumerate(colors_rgb):
    hsv = colorsys.rgb_to_hsv(r=r, g=g, b=b)
    
    colors_hsv.append(hsv)

colors_hsv = np.array(colors_hsv)
colors_hsv[:, :2] *= 255
colors_hsv = np.rint(colors_hsv).astype(np.uint8)
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

```python
new_img_color_hsv_idx = pairwise_distances_argmin(img_hsv.reshape((-1, 3))[:, :2], np.array(colors_hsv)[:, :2]).reshape(img.shape[:2])

new_img_hsv = np.zeros_like(img, dtype=np.uint8)

for y in range(img.shape[0]):
    for x in range(img.shape[1]):
        color = np.array(colors_hsv[new_img_color_hsv_idx[y, x]], np.uint8)

        color[2] = img_hsv[y, x, 2]
        
        new_img_hsv[y, x] = color
        
plt.imshow(cv2.cvtColor(new_img_hsv, cv2.COLOR_HSV2RGB))
```

```python
plt.figure(figsize=(16, 10))
plt.subplot(1, 3, 1)
plt.imshow(img)
plt.subplot(1, 3, 2)
plt.imshow(new_img)
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(new_img_hsv, cv2.COLOR_HSV2RGB))
```
